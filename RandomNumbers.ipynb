## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import median
```

## Generate Random numbers between 1 and 100

```python
randnum = np.random.uniform(low=0, high=100, size=10000)
```

### Plotting Frequency Graphs

```python
def plottingFreqGraph(list_name, lower, upper, binSize):
    numberOfBins = int((upper-lower)/binSize)
    plt.hist(list_name, bins=numberOfBins, rwidth=0.9,color='#607c8e')
    plt.title('Frequency Distribution with binSize: {}'.format(binSize))
    plt.xlabel('Numbers')
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
```

```python
plottingFreqGraph(randnum, 0, 100, 1)
```

```python
plottingFreqGraph(randnum, 0, 100, 5)
```

```python
plottingFreqGraph(randnum, 0, 100, 10)
```

### Generating normally distributed random numbers and graph plotting

```python
randnum_norm = np.random.normal(50, 10, size=10000)
```

```python
plottingFreqGraph(randnum_norm, 0, 100, 5)
```

### Calculating mean, median and std deviation of normally distributed random numbers

```python
#calculating mean#
def calculateAverage(list_numbers):
    sum=0
    for num in list_numbers:
        sum = sum + num
    mean= sum/len(list_numbers)
    return mean


    
calculateAverage(randnum_norm)
```

```python
#calculating median#
median(randnum_norm)
```

```python
#calculating std deviation #
def calculateStdDev(list_numbers):
    sum = 0
    avg = calculateAverage(list_numbers)
    for number in list_numbers:
        sum = sum + (number - avg)**2
    stddev = (sum/len(list_numbers))**0.5
    return stddev
```

```python
calculateStdDev(randnum_norm)
```

```python
import statistics
statistics.stdev(randnum_norm)
```

### Calculating  percentile on random normally distributed numbers

```python
#calculating percentile#
def calculatePercentile(list_numbers, percent):
    percentile_value = np.percentile(list_numbers, percent)
    return percentile_value

calculatePercentile(randnum_norm, 75)
```

### Calculating datapoints lie between (mean-stddev,mean+stddev)

```python
def countDataPoints(list_numbers):
    avg = calculateAverage(list_numbers)
    stdDev = calculateStdDev(list_numbers)
    low = avg - stdDev
    high = avg + stdDev
    counter = 0
    for number in list_numbers:
        if ((number > low.item()) and (number < high.item())):
            counter = counter + 1
    return counter
```

```python
countDataPoints(randnum_norm)
```
